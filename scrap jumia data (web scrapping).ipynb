```python
from bs4 import BeautifulSoup  
import requests
import pandas as pd
import numpy as np
```

```python
def get_title(soup):
    try:
        title = soup.find("h1", attrs={'class': '-fs20 -pts -pbxs'})
        title_value = title.text
        title_string = title_value.strip()
    except AttributeError:
        title_string = ' '
    return title_string

def get_price(soup):
    try:
        price = soup.find("span", attrs={'class': '-b -ltr -tal -fs24 -prxs'}).text.strip()
    except AttributeError:
        price = ' '
    return price

def get_rating(soup):
    try:
        rating = new_soup.find("div", attrs={'class': 'stars _m _al'}).text.strip()
    except AttributeError:
        rating = ' '
    return rating

def get_reviews(soup):
    try:
        articles = soup.find_all("article", class_="-pvs -hr _bet")
        reviews_list = []
        for article in articles:
            title = article.find("h3").text.strip()
            comment = article.find("p").text.strip()
            reviews_list.append(f"{title}: {comment}")
        reviews = '\n'.join(reviews_list)
    except AttributeError:
        reviews = ' '
    return reviews

def get_availability(soup):
    try:
        availability = soup.find("p", attrs={'class': '-df -i-ctr -fs12 -pbs -rd5'}).text.strip()
    except AttributeError:
        availability = 'Not available'
    return availability
```

```python
if __name__ == '__main__':
    HEADERS = ({'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36',
           'Accept-Language': 'en-US, en;q=0.5'})
    URL = "https://www.jumia.ma/catalog/?q=playstation+4"
    webpage = requests.get(URL, headers=HEADERS)
    soup = BeautifulSoup(webpage.content, "html.parser")
    links = soup.find_all("a", attrs={'class': 'core'})
    links_list = []
    for link in links:
        links_list.append(link.get('href'))
    d = {'title':[], 'price':[], 'rating':[], 'reviews':[], 'availability':[]}
    for link in links_list:
        new_webpage = requests.get("https://www.jumia.ma" + link, headers=HEADERS)
        new_soup = BeautifulSoup(new_webpage.content, "html.parser")

        d['title'].append(get_title(new_soup))
        d['price'].append(get_price(new_soup))
        d['rating'].append(get_rating(new_soup))
        d['reviews'].append(get_reviews(new_soup))
        d['availability'].append(get_availability(new_soup))

    jumia_df = pd.DataFrame.from_dict(d)
    jumia_df['title'].replace('', np.nan, inplace=True)
    jumia_df = jumia_df.dropna(subset=['title'])
    jumia_df.to_csv("jumia_data.csv", header=True, index=False)
```
